```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Potential Equation
# H = U - k*cos(z)
t = np.arange(1.5*np.pi,2.5*np.pi,0.01)   # start,stop,step
#z =  - 5*np.cos(t) - 0.8*np.cos(t)*np.cos(16*t) + 0.8
#z = - 5*np.sin(t) + np.sin(t)*2.5*(1 - np.cos(16*t))
y = a*(1-np.cos(t))
z =   a*(1-np.cos(t)) + f*a*(1-np.cos(t)*np.cos(16*t))
```

```python
# Now, we define a function that returns ˙z and ¨z (in that order).
```

```python
#def z_derivatives(x, t):
#    return [x[1], -(1/x[0])*(x[1]**2 + b*x[1] + g*x[0] - g*h)]
#    return [x[1], -a*x[0]]
#   return  [x[1], -a*np.sin(2*3.14*x[0]/180)-b*n*np.sin(2*3.14*n*x[0]/180)]
def mind_derivatives(x, t):
    return  [x[1], -a*np.sin(2*3.14*x[0]/180) -b*n*np.sin(2*3.14*n*x[0]/180)]
```

```python
# Next, we create an array of the desired times for the solution.
time = np.arange(0, 500, 10e-2)
```

```python
position, velocity = odeint(mind_derivatives, [2e-3, 0], time).T
```

```python
# A plot of the fluid height (position) inside the straw as a function of time.
plt.plot(time, position*100, 'orange', linewidth = 2)
plt.xlabel('time (s)')
plt.ylabel('position (cm)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.grid(True)
```
