# Intellect Potential Well of Mind

```python
I=1.0
a=4.0
b=0.1
c=1.5
f=0.3  # it should be a fraction
N=18 # No of anaisotropy in the potential well = 1+N/2
k=1.38 # boltzmann constant, actual value is 1.380649E-23 Joules/Kelin
T=1.7    # at T=1.7 tou=5, E/kT > 1
print("kT :", k*T)
```

```python
# Mind Potential Equation biased because of  various brands in the market
t = np.arange(-np.pi/2,np.pi/2,0.01)   # start,stop,step
omega=0.5
# New Potential
H0 = 0.5*I*omega*omega + a*(1-np.cos(t))
H1 = 0.5*I*omega*omega + a*(1-np.cos(t)) + f*a*(1-np.cos(N*t))
H2 = 0.5*I*omega*omega + a*(1-np.cos(t)) + f*a*(1-np.cos(N*t)) + 0.5*c*omega*omega

# A plot of the fluid height (position) inside the straw as a function of time.
plt.title("Potential Well at different situation")
plt.margins(x=0, y=0)
plt.ticklabel_format(style='plain', axis='x')
plt.plot(t*180/np.pi, H0, t*180/np.pi, H1, t*180/np.pi, H2, 'black', linewidth = 1)
plt.xlabel('State of Mind (theta)')
plt.ylabel('Free Energy well (H)')
plt.legend(['Normal Mind', 'Biased Mind', 'External Influenced Mind'])
plt.grid(False)
```

```python
def find_maxima_minima(series):
    maxima = []
    minima = []
    max_min = []
    for i in range(1, len(series) - 1):
        if series[i] > series[i - 1] and series[i] > series[i + 1]:
            maxima.append((-np.pi/2 + np.pi*i/315 , series[i]))
            max_min.append((-np.pi/2 + np.pi*i/315 , series[i]))
        elif series[i] < series[i - 1] and series[i] < series[i + 1]:
            minima.append((-np.pi/2 + np.pi*i/315, series[i] ))
            max_min.append((-np.pi/2 + np.pi*i/315 , series[i]))
    return maxima, minima, max_min

local_max, local_min, local_max_min = find_maxima_minima(H2)
print("Length of data",len(H2))
print("[angle,energy]")
print(local_max)
print(local_min)
print(local_max_min)
print("Length of data",len(local_max_min))
total_max_min = len(local_max_min)
```

# Dynamics of Pure Mind: Physical Pendulum

```python
def pure_mind(x, t):
    if   x[0] > 0.5*np.pi-.06:
        x[0]=1.0*np.pi - x[0] +0.06 # oscillate between -pi/2 to +pi/2
    elif x[0]< -0.5*np.pi+0.06:
        x[0]=-x[0] - 1.0*np.pi -0.06 # oscillate between -pi/2 to +pi/2
    return  [x[1], -a*np.sin(x[0]) ]
```

```python
# Next, we create an array of the desired times for the solution.
# Create an empty NumPy array of zeros
no_of_samples =400
time_array = np.zeros(no_of_samples)
posi_array = np.zeros(no_of_samples)  # displacement or position
velo_array = np.zeros(no_of_samples)  # velocity
KE_array = np.ones(no_of_samples)  # Kinetic energy
PE_array = np.ones(no_of_samples)  # Potential energy
my_time = [0,1e-2] # odeint takes time in terms of array only, h=t1-t0
i=0
theta = np.pi/6 -.001
omega =0

my_time = [0,1e-2] # odeint takes time in terms of array only, h=t1-t0
init_value =[theta, omega] # odeint takes init_value in terms of array only

KE_array[0]= 0.5*I*omega*omega #+ a*(1-np.cos(theta))
PE_array[0]= a*(1-np.cos(theta))
while i != no_of_samples:
    solution = odeint(pure_mind, init_value, my_time)
    posi_array[i]=solution[1,0]
    velo_array[i]=solution[1,1]
    time_array[i]=i
    theta=solution[1,0]
    omega=solution[1,1]
    init_value=[theta, omega]
    KE_array[i]= 0.5*I*omega*omega
    PE_array[i]= a*(1-np.cos(theta))
    i=i+1

from matplotlib import colors
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (sample number)')
ax1.set_ylabel('angular displacement(degree) and velocity (degree/sample)', color='black')
ax1.plot(time_array[0:5000], 180*posi_array[0:5000]/3.14,'orange', time_array[0:5000], 180*velo_array[0:5000]/3.14,  'blue', linewidth = 2.5)
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(['Angular Displacement', 'Angular Speed'])
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Energy', color='red')  # we already handled the x-label with ax1
ax2.plot(time_array[0:5000], KE_array[0:5000],'red',time_array[0:5000], PE_array[0:5000],'green',  linewidth = 1.0)
ax2.tick_params(axis='y', labelcolor='black')
ax2.legend(['Kinetic Energy','Potential Energy'])
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
# Initialise the subplot function using number of rows and columns
figure, axis = plt.subplots(2, 2)

# For Sine Function
axis[0, 0].plot(time_array,PE_array,'green', linewidth = 1.5)
axis[0, 0].legend(['Potential Energy vs. sample no.'],fontsize="8", loc ="upper left")

# For Cosine Function
axis[0, 1].plot(time_array, KE_array,'red', linewidth = 1.5)
axis[0, 1].legend(['Kinetic Energy vs. sample no.'],fontsize="8", loc ="upper left")

# For Tangent Function
axis[1, 0].plot(time_array, 180*posi_array/3.14,'orange', linewidth = 1.5)
axis[1, 0].legend(['Angular Disp. vs. sample no.'],fontsize="8", loc ="upper left")

# For Tanh Function
axis[1, 1].plot(time_array, 180*velo_array/3.14,'blue', linewidth = 1.5)
axis[1, 1].legend(['Angular Velocity vs. sample no.'],fontsize="8", loc ="upper left")

# Combine all the operations and display
plt.show()
```

# True Mind oscillator

```python
def mind_osc(x, t):
    if   x[0] > 0.5*np.pi-.06:
        x[0]=1.0*np.pi - x[0] +0.06 # oscillate between -pi/2 to +pi/2
    elif x[0]< -0.5*np.pi+0.06:
        x[0]=-x[0] - 1.0*np.pi -0.06 # oscillate between -pi/2 to +pi/2
    return  [x[1], -a*np.sin(x[0]) -b*x[1] - f*a*N*np.sin(N*x[0]) + c*x[1]]
```

```python
# Next, we create an array of the desired times for the solution.
# Create an empty NumPy array of zeros
no_of_samples =130000
time_array = np.zeros(no_of_samples)
posi_array = np.zeros(no_of_samples)  # displacement or position
velo_array = np.zeros(no_of_samples)  # velocity
enrg_array = np.ones(no_of_samples)  # energy
KE_array = np.ones(no_of_samples)  # Kinetic energy
PE_array = np.ones(no_of_samples)  # Potential energy

I=1.0
a=4.0
b=0.8
c=0.05
f=0.3  # it should be a fraction
N=18 # No of anaisotropy in the potential well = 1+N/2
k=1.38 # boltzmann constant, actual value is 1.380649E-23 Joules/Kelin
T=1.7    # at T=1.7 tou=5, E/kT > 1
print("kT :", k*T)

theta = np.pi/3 -.001
omega =0
tou_0=40
tou=tou_0

my_time = [0,1e-2] # odeint takes time in terms of array only, h=t1-t0
init_value =[theta, omega] # odeint takes init_value in terms of array only
KE_array[0]=0.5*I*omega*omega
PE_array[0]=a*(1-np.cos(theta))+ f*a*(1-np.cos(N*theta))
enrg_array[0]= 0.5*I*omega*omega + a*(1-np.cos(theta))+ f*a*(1-np.cos(N*theta)) + 0.5*c*omega*omega

flip_elapsed_time = 50 #int(tou)
print("energy, tau", enrg_array[0], tou)
print("E/kT:",enrg_array[0]/k*T )
i=0
E_high = 0
E_low  = 0

while i != no_of_samples:
    for m in range(0,int((total_max_min/2)),1):
        t1,e1 = local_max_min[2*m] # this is minima
        t2,e2 = local_max_min[2*m+1] # this is maxima
        if theta >=t1 and theta <= t2:
            tou=tou_0*np.exp(abs(e2-e1)/(k*T))
            #print("theta",theta, 180*t1/3.14,180*t2/3.14)
            #print("enrg", e1, e2)
            #print("tou", tou)

    solution = odeint(mind_osc, init_value, my_time)
    posi_array[i]=solution[1,0]
    velo_array[i]=solution[1,1]
    time_array[i]=i
    theta=solution[1,0]
    omega=solution[1,1]
    init_value=[theta, omega]
    KE_array[i]=0.5*I*omega*omega
    PE_array[i]=a*(1-np.cos(theta))+ f*a*(1-np.cos(N*theta))
    enrg_array[i]= 0.5*I*omega*omega + a*(1-np.cos(theta)) + f*a*(1-np.cos(N*theta)) + 0.5*c*omega*omega
    if velo_array[i-1] > velo_array[i - 2] and velo_array[i-1] > velo_array[i] and  abs(velo_array[i-1])<= 0.25:
        E_low = 1 # latch
        #print(c,i,"For E_low ",velo_array[i-2]," ",flip_elapsed_time)

    if velo_array[i-1] > velo_array[i - 2] and velo_array[i-1] > velo_array[i] and abs(velo_array[i-1]) > 0.25:
        E_high  = 1 # latch
        #print(c,i,"For E_high ",velo_array[i-2]," ",flip_elapsed_time)
    ############################################################
    if   flip_elapsed_time < tou and E_high==1:
        c=0.001 # normal operation
    elif flip_elapsed_time < tou and E_low ==1:
        c=0.535*np.random.random(1)[0]   # knowledge boost
        omega = math.sqrt(I+c)*omega
        #omega=omega+ 0.3 # this works well
        init_value=[theta, omega]
    elif flip_elapsed_time >= tou:
        flip_elapsed_time=0
        E_high = 0
        E_low  = 0

    flip_elapsed_time = flip_elapsed_time + 1 # this is under for loop
    i=i+1 # this is under for loop

print("End of while loop")

```
